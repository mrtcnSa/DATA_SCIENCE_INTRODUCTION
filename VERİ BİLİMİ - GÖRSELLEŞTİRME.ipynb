#  Veri Görselleştirme Nedir?

Bugün iş dünyasında veri analizi yoluyla toplanan bu kadar çok bilgi ile, bu verilerin bir resmini çizmenin bir yolunu bulmalıyız, böylece onları yorumlayabiliriz.

Veri görselleştirme, haritalara veya grafiklere görsel bağlam sağlayarak bilginin ne anlama geldiğine dair bize net bir fikir verir.
Bu, verileri insan zihninin anlaması için daha doğal hale getirir ve bu nedenle büyük veri kümelerindeki eğilimleri, modelleri ve aykırı değerleri tanımlamayı kolaylaştırır.

Kısaca, veri görselleştirme verilerin veya bilgilerin bir grafik, çizelge veya diğer görsel formatta temsilidir.


## Veri Görselleştirmenin Önemi
Ana amaç, veri ve istatistikleri görselleştirmek, bilgi edinmek için ekranları yorumlamaktır. Veri görselleştirme, veri temizleme, veri yapısını keşfetme, aykırı değerleri ve olağan dışı grupları tespit etme, eğilimleri ve kümeleri belirleme, yerel kalıpları belirleme, modelleme çıktısını değerlendirme ve sonuçları sunma için kullanışlıdır.


Makine öğrenimi, tahmine dayalı analiz gibi analizler yapmayı kolaylaştırır ve bu analizler, daha sonra sunum için yararlı görselleştirmeler işlevi görebilir.
Finans, pazarlama, teknoloji, tasarım veya başka herhangi bir alanda çalışıyor olun, verileri görselleştirmeniz gerekir. Bu gerçek, veri görselleştirmenin önemini göstermektedir.

```python
mu1 = 100
sigma1 = 15
x = mu1 + (sigma1 * np.random.randn(10000))

mu2 = 110
sigma2 = 10
y = mu2 + (sigma2 * np.random.randn(10000))

# söz konusu veriLerin histogramı

n, bins, patches = plt.hist(x, 100, density = 1, facecolor='b', alpha=0.25)
n, bins, patches = plt.hist(y, 100, density = 1, facecolor='g', alpha=0.25)

plt.xlabel('Veriler')
plt.ylabel('Olasılıklar')
plt.title(r'Verilerin Histogrami: $\mu_1=100,\ \mu_2=110$')
plt.annotate('Gauss eğrisinin tepesi', xy=(110,0.04), xytext=(130,0.045), arrowprops=dict(facecolor='black', shrink=0.1))

plt.text(75, 0.025, r'$\mu_1=100,\ \sigma_1=15$') 
plt.text(120, 0.035, r'$\mu_2=110,\ \sigma_2=10$')
plt.axis([40, 160, 0, 0.05])
plt.grid(True)
plt.show()
```

## Histogram
• Histogram, gruplandırılmış bir veri dağılımının sütun grafiğiyle gösterimidir.

***
Veri dağılımını görselleştirmek için sıklık dağılımı grafiği olarak kullanılır. Bir histogram, x-ekseni boyunca veri aralıklarının (bölümlerin) yer aldığı bir çubuk grafiğidir. Y ekseninde ise her bir bölümde kaç adet veri örneği olduğunu gösteren sıklık sayısı yer alır.
***
• Tekrarlı sayılardan oluşan verilerin, uygulanan işlemlerden sonra önce tabloya, tablodan yararlanarak grafiğe aktarılması, yani veri gruplarının grafiğinin dikdörtgen sütunlar halinde gösterilmesidir.

```python
sns.displot(data,x="kilo")
plt.xlabel("Kilo")
plt.ylabel("Frekans")
plt.title("Ağırlık Histogramı")
plt.show()
```

### Histogtram - Kategorik

Bu tür kategorik değişkenler için veri dağılımını göstermek için, bar grafiği veya pie chart (pasta grafiği) kullanmak daha yaygın olarak tercih edilir.

##  2 Boyutlu Histogram

2 farklı değişkenin birbirleriyle olan ilişkisini görselleştirmek için kullanılır. Bu histogram, iki boyutlu bir histogramdaki her bir kareye karşılık gelen bir renk tonu kullanır. Renk tonu, karenin içindeki noktaların sayısına göre belirlenir.

```python
sns.displot(data, x="kilo", y="boy", kind="kde")
plt.xlabel("Ağırlık")
plt.ylabel("Boy")
plt.title("Ağırlık-Boy Histogramı")
plt.show()
```

kind="kde" parametresi, seaborn kütüphanesindeki distplot() fonksiyonunda kullanılır ve bu parametre kullanıldığında, bir yoğunluk grafiği oluşturulur. Yoğunluk grafiği, bir veri kümesindeki yoğunluğun nasıl dağıldığını gösterir.

```python
sns.displot(data, x="kilo", y="boy", hue='cinsiyet', kind="kde")
plt.xlabel("Ağırlık")
plt.ylabel("Boy")
plt.title("Cinsiyete göre ayrıştırılmış Ağırlık-Boy Histogramı")
plt.show()
```

• Plotly, verileri yükleyerek ve grafiksel kullanıcı arayüzü kullanarak etkileşimli grafikler oluşturmak için teknik geçmişi olmayan kişiler tarafından da kullanılabilir.

***
Plotly, Python dilinde kullanılan bir görselleştirme kütüphanesidir. Plotly, interaktif ve özelleştirilebilir grafikler oluşturmak için kullanılır. Bu kütüphane, çizgi grafikleri, bar grafikleri, scatter grafikleri, yüzey grafikleri, harita grafikleri ve daha birçok farklı grafik türünü destekler.

Plotly, Python'da pandas, numpy, matplotlib ve seaborn gibi diğer veri işleme ve görselleştirme kütüphaneleri ile entegre edilebilir. Plotly, grafiklerinizi Python'daki diğer kütüphanelerle birlikte kullanarak daha fazla analiz yapmanıza ve verilerinizi daha iyi anlamanıza olanak tanır.

Plotly aynı zamanda bir bulut tabanlı veri analizi platformudur. Bu platform, veri bilimcilerin, araştırmacıların ve işletme analistlerinin verilerini görselleştirmelerine, paylaşmalarına ve işbirliği yapmalarına olanak tanır. Kategorik verileri temsil ettikleri değerlerle orantılı yükseklik veya uzunluk özetlemek

# Dairsel Grafik
• Dairesel grafik, istatistik biliminde betimsel istatistik alanında kategorik verileri görsel bir şekilde betimleyip özetlemek için hazırlanan; içindeki kategori dilimlerini orantısal olarak gösteren bir daire şeklinde sunulan bir gösterim aracıdır.

```python
veri2011 = rateData[rateData.yil == 2011].iloc[:8,:]
dilim1 = veri2011.mezun_istihdamı
dilim1_liste  = [float(each.replace(',','.')) for each in veri2011.üniversite ]
etiketler = veri2011.fakülte_kalitesi
                 
                 
                 
cizgi = go.Pie(labels = etiketler,
               values = dilim1_liste,
                hoverinfo = 'label+value+percent',
                textinfo = 'value+percent',
                textfont = dict(size=8),
                rotation = 180,
                hole = 0.3,
                marker = dict(line=dict(color='#000000', width=1)))
veri_ = [cizgi]
yerlesim = dict(title = '2011 Yılı - İlk 8 Üniversitenin Öğrenci Sayıları' ,legend = dict(orientation="h"))
fig = dict(data=veri_, layout=yerlesim)
plot(fig, filename = '3_dairesel-öğrenci sayısı.html')
```
